```python
xpoints = np.array([0, 10, 20, 40])
ypoints = np.array([10, 50, 60, 70])

xpoints2 = np.array([1, 4, 6, 40])
ypoints2 =  np.array([2, 10, 17, 20])
plt.plot(xpoints, ypoints, '*', linestyle = 'dotted', label='line1')
plt.plot(xpoints2, ypoints2, label='line2')
plt.legend()
plt.show()
```
