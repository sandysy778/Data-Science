Gaussian Filter

```python
doc_image = cv2.imread('passport.jpeg')
```

```python
cv2.imshow("My Image", doc_image)
cv2.waitKey()
```

```python
gray_doc_img = cv2.cvtColor(doc_image, cv2.COLOR_BGR2GRAY)
```

```python
gray_doc_img
```

Gaussian Filter

```python
gau_doc = cv2.GaussianBlur(doc_image, (3, 3), cv2.BORDER_DEFAULT)
```

```python
cv2.imshow("My Image", gray_doc_img)
cv2.imshow("GaussianBlur Image", gau_doc)

cv2.waitKey()
```

```python
#Image
#min value
#max value
edge_image = cv2.Canny(image, 10, 15)
```

```python
cv2.imshow("My Image", image)
cv2.imshow("GaussianBlur Image", edge_image)

cv2.waitKey()
```
