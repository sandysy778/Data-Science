```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
oh = OneHotEncoder(handle_unknown='ignore',drop= "first")
```

```python
#Init the ColunmTransformer

#transformers
#list of tuples
#[(), ()], (name, object/instance, column's)
#reminder
#drop, passthrough

ct = ColumnTransformer(transformers = [("ohe", oh,["State"])], remainder = "passthrough") 

```

```python
x = ct.fit_transform(x)
```

```python
ct.get_feature_names_out()
```

```python
column_names = []
for each_col in ct.get_feature_names_out():
    column_names.append(each_col.split("_")[-1])
```

```python
column_names
```

```python
X = pd.DataFrame(x)
```

```python
X.columns = column_names
```

```python
X
```

```python
y = dataset["Profit"]
```

Split data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

Building the Decision Tree  Model   

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt_model = DecisionTreeRegressor(criterion = "squared_error")  
```

```python
dt_model.fit(X_train, y_train)
```

Prediction

```python
y_pred = dt_model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
y_pred_train = dt_model.predict(X_train)
```

Evaluation
