# Filling Missing Values(imputation)

```python
import numpy as np
d1 = {'score1': [100,80, np.nan, 95],
      'score2': [30, 20, 56, np.nan],
      'score3': [np.nan, 40, 60, 60]

     }
df1 = pd.DataFrame(d1)
df1
```

```python
df1.fillna(1000)
```

```python
df1.fillna(
    {
      'score1': 1000,
      'score2': 2000,
      'score3': 3000
    }


)
```
