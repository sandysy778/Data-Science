```python
dtype_ch={"Storage":"int","Battery":"int"}
```

```python
def change_dtype(df,dtype_ch):
    df_copy=df.copy()
    for col,dtype in dtype_ch.items():
        new_df=df_copy[col]=df_copy[col].astype(dtype)
    return new_df
        
        
```

```python
def numerical_univariate(df):
    """This function will take only numerical columns
    and give you univariate analysis of those columns"""

    df_copy=df.copy()
    
    for i in df.select_dtypes(include="number").columns:
        
        plt.figure(figsize=(10,5))
        plt.hist(df[i],bins=20)
        plt.title(f"HISTOGRAM-{i}")
        plt.show()
        
        plt.figure(figsize=(10,5))
        sns.boxplot(df[i])
        plt.title(f"BOXPLOT- {i}")
        plt.show()
        
        des=df[i].describe()
        print(des)
        
        skew=df[i].skew()
        print(skew)
        
        k=df[i].kurtosis()
        print(k)
        
        print("*************************************************************")
        print()
        
```

```python
numerical_univariate(df)
```

```python
ax=df.groupby("Brand")["Price"].agg("min").plot(kind="bar")

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=6, color='black')

plt.tight_layout()
plt.show()
```
