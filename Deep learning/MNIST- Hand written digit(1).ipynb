```python
#tqdm -- to show the progress

#X = [Img_array1, Img_array2,Img_array3....Img_arrayn]
#y = ["2","5","3","4"]
X =[] #Input or features -- pixels values
y= [] #target -- labels

#crawling to all the folders
for each_folder in tqdm.tqdm(folder_paths):
    # get the classname from folder_path
    class_name = each_folder.split("\\")[-1]
    #crawling to all the image in the each_folder
    for img_path in glob.glob(each_folder + "\*.jpg"):
        #print(img_path)
        #Read the image from the path
        img = Image.open(img_path)
        img = img.convert("L")
        #20% of image will have different mage 
        if np.random.randint(1,10) <= 2:
            thresh = 100
            #point will help to apply a function on each and every pixel value
            img = img.point(lambda p:0 if p > thresh else 255)
        
        #Convert to numpy array
        img_array = np.array(img)
        
            
        #flatten to covert 28,28 to 784 pixels to form a row
        img_array = img_array.flatten()
        #append the img in X
        X.append(img_array)
        #append the class_name
        y.append(class_name)
```

### Split into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, random_state = 35)
```

```python
from tensorflow import keras
```

```python
y_train
```

```python
y_train = keras.utils.to_categorical(y_train)
```

```python
y_test = keras.utils.to_categorical(y_test)
```

```python
y_train.shape
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = sc.transform(X_test)
```

### Build The model

```python
#Sequential -- init the Neural Networ model
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input
```

Init the Neural Network model

```python
nn_model = Sequential()
```

```python
X_train.shape
```

Add input layer

```python
X_train.shape[1]
```

```python
#shape -- shape of input, no of nerons in input layer
inp_layer = Input(shape = (X_train.shape[1],) )
```

```python
nn_model.add(inp_layer)
```

Add Hidden layer

```python
#units -- -- no of nerons in hidden Layer
#kernel --- weights-- kerenal_initializer -- strategy to init the weight
#activation -- the activation function for this Layer
hidden_layer = Dense(units = 6, kernel_initializer = "random_uniform", activation = "relu", bias_initializer = "random_uniform")
```

```python
nn_model.add(hidden_layer)
```

Add Output layer

```python
#units -- no of neurons in output layer
#kernel -- weights -- kernel_initialzer --- strategy to init the weights
#activation -- the activation function for  this layer 
output_layer = Dense(units = 1, kernel_initializer = "random_uniform", activation = "softmax", bias_initializer = "random_uniform")
```

```python
nn_model.add(output_layer)
```

```python
nn_model.summary()
```

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(nn_model, show_shapes = True, show_dtype = True)
```

```python
#compile 
#strategies to be followed while training

nn_model.compile(loss = "categorical_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```python
nn_model.fit(X_train, y_train, epochs = 5, batch_size = 32)
```
