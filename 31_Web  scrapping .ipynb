```python
for page in range(1,62):
    URL = 'https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
    #print(URL)
    
    htmlpage = requests.get(URL)
    soup = BeautifulSoup(htmlpage.text)
    
    search_laptops = soup.find_all('div', attrs = {'class': '_3pLy-c row'})
    
    prices = []
    ratings = []
    features = []
    titles = []
    for laptop in search_laptops:
        
        p = laptop.find('div', attrs = {'class': '_30jeq3 _1_WHN1'})
        if(p is None):
            prices.append(np.nan)
        else:
            prices.append(p.text)
            
            
        p = laptop.find('div', attrs = {'class': '_3LWZlK'})
        if(p is None):
            ratings.append(np.nan)
        else:
            ratings.append(p.text)
            
            
        p = laptop.find('div', attrs = {'class': 'fMghEO'})
        if(p is None):
            features.append(np.nan)
        else:
            features.append(p.text)
            
            
        p = laptop.find('div', attrs = {'class': '_4rR01T'})
        if(p is None):
            titles.append(np.nan)
        else:
            titles.append(p.text)
    
```

```python
len(prices), len(ratings),len(features), len(titles)
```

```python
laptop_df = pd.DataFrame({'Title':titles,
                         'price': prices,
                          'Ratings': ratings,
                          'features': features})
laptop_df.head()
```
